# Regression
## Multiple Linear Regression

$\begin{equation}
    y = \beta_0 + \boldsymbol{\beta} \boldsymbol{x}
\end{equation}$

## Logistict Regression

$
\begin{equation}
    y = \frac{1}{1+e^{-(\beta_0 + \boldsymbol{\beta} \boldsymbol{x}})}
\end{equation}$

## Sum of Squares
$
\begin{equation}
    Total\_{SS} = \Sigma_{i=1}^n(y_i - \overline{y})^2
\end{equation}$

$
\begin{equation}
    Regression\_{SS} = \Sigma_{i=1}^n(\hat{y_i} - \overline{y})^2
\end{equation}$

$
\begin{equation}
    Residual\_{SS} = \Sigma_{i=1}^n(y_i - \hat{y_i})^2
\end{equation}$

$
\begin{equation}
    Total\_{SS} = Regression\_{SS} + Residual\_{SS}
\end{equation}$

## $R^2$

$
\begin{equation}
    R^2 = 1 - Residual\_SS / Total\_SS
\end{equation}$

$
\begin{equation}
    Adj\_R^2 = 1 - \frac{Residual\_SS/(n-K)}{Total\_SS/(n-1)}
\end{equation}$

where K is number of predictor parameters and n is sample size.

## Regularization
$
\begin{equation}
    Lasso(\hat{\beta}) = \Sigma_{i=1}^n(y_i - x_i \hat{\beta})^2+ \lambda \Sigma_{j=1}^m |\hat{\beta}_j|
\end{equation}$

$
\begin{equation}
    Ridge(\hat{\beta}) = \Sigma_{i=1}^n(y_i - x_i \hat{\beta})^2 + \lambda \Sigma_{j=1}^m w_j \hat{\beta}_j^2
\end{equation}$

# Relations

## Standard error
$
\begin{equation}
    SE = \frac{s}{\sqrt{n}}
\end{equation}$

## Variance
$
\begin{equation}
    var(N(s))=E[N(s)^2]-E[N(s)]^2
\end{equation}$

## Sample Variance
$
\begin{equation}
    s^2=\frac{1}{n-1}\Sigma_{i=1}^n(y_i - \overline{y})^2
\end{equation}$
where s is standard deviation.

## Covariance
$
\begin{equation}
    cov(N(s),N(t))=E[N(s)N(t)]-E[N(s)]E[N(t)]
\end{equation}$

## Sample Covariance
$
\begin{equation}
    cov_{x,y}=\frac{\Sigma_{i=1}^n(x_i-\overline{x})(y_i-\overline{y})}{n-1}
\end{equation}$

## Correlation
$
\begin{equation}
    cor(N(s),N(t))=\frac{cov(N(s),N(t))}{\sqrt{var(N(s))var(N(t))}}
\end{equation}$

## Sample Correlation
$
\begin{equation}
    corr_{x,y}=\frac{\Sigma_{i=1}^n(x_i-\overline{x})(y_i-\overline{y})}{\sqrt{\Sigma_{i=1}^n(x_i-\overline{x})^2\Sigma_{i=1}^n(y_i-\overline{y})^2}}
\end{equation}$

# Distributions

## Discrete

### Binomial Distribution
$
\begin{equation}
    f(x)=\binom{n}{x} p^x (1-p)^{n-x}
\end{equation}$

$
\begin{equation}
    \binom{n}{x}=\frac{n!}{(n-x)!x!}
\end{equation}$

## Continuous

### Normal Distribution
$
\begin{equation}
    f(x)=\frac{1}{\sqrt{2 \pi \sigma^2}} e^{\frac{-(x - \mu )^2}{2 \sigma^2}}
\end{equation}$

### Exponential Distribution
$
\begin{equation}
    f(x)=\lambda e^{-\lambda x}
\end{equation}$
$
\begin{equation}
    Q(t)=e^{-\lambda t}
\end{equation}$
$
\begin{equation}
    f^*(s)=\int_0^{\infty}e^{-st}\lambda e^{-\lambda t}dt=\frac{\lambda}{\lambda + s}
\end{equation}$

### Gamma Distribution
$
\begin{equation}
\Gamma (1)=1
\end{equation}$
$
\begin{equation}
\Gamma (n+1)=n!
\end{equation}$
$
\begin{equation}
\Gamma (1/2)=\sqrt{\pi}
\end{equation}$
$
\begin{equation}
    f^*(s)=\int_0^{\infty}e^{-st}\frac{\beta^{\alpha t^{\alpha - 1}e^{-\beta t}}}{\Gamma (\alpha)}dt=(\frac{\beta}{\beta + s})^{\alpha}
\end{equation}$

### Substitution

Example:
$
\begin{equation}
    \int \lambda e^{-\lambda x}dx
\end{equation}$

Set substitution:
$
\begin{equation}
    u = -\lambda x
\end{equation}$

$
\begin{equation}
    \frac{du}{dx} = -\lambda
\end{equation}$

$
\begin{equation}
    -\frac{du}{\lambda} = dx
\end{equation}$

Resolve:
$
\begin{equation}
    - \frac{1}{\lambda} \int \lambda e^{u}du = - \int e^{u}du =  - e^{u} = - e^{-\lambda x}
\end{equation}$

### By Parts

Formula:
$
\begin{equation}
\int f g' = fg - \int f' g
\end{equation}$
%
\\
Example:
$
\begin{equation}
    \int x e^{-\lambda x}dx
\end{equation}$

Set parts:
$
\begin{equation}
    f = x
\end{equation}$
$
\begin{equation}
    f' = 1
\end{equation}$
$
\begin{equation}
    g' = e^{-\lambda x}
\end{equation}$
$
\begin{equation}
    g = - \frac{e^{-\lambda x}}{\lambda}
\end{equation}$

Resolve:
$
\begin{equation}
    - x \frac{e^{-\lambda x}}{\lambda} + \int \frac{e^{-\lambda x}}{\lambda} = - \frac{x e^{-\lambda x}}{\lambda} - \frac{e^{-\lambda x}}{\lambda^2}
\end{equation}$

# Laplace Transform
Linearity:
$
\begin{equation}
h(t)=af(t)+bg(t)
\end{equation}$
$
\begin{equation}
h^*(s)=af^*(s)+bg^*(s)
\end{equation}$
\\
Special cases:\\
$f^*(0)=1$\\
If $f(t)=1$, then $f^*(s)=1$.\\
If $f(t)=t$, then $f^*(s)=1/s^2$.
\\
Derivative:
$\begin{equation}
f'^*(s)=sf^*(s)-f(0)
\end{equation}$
\\
Special cases:\\
$f^*(0)=1$\\
\\
Integration:
$\begin{equation}
F^*(s)=\frac{f^*(s)}{s}
\end{equation}$
\\
Special cases:\\
$F(0)=0$\\
\\
Convolution:
$\begin{equation}
    h^*(s)=f^*(s)g^*(s)
\end{equation}$
Special cases:\\
If f and g are pdfs of two independent variables T and U, then:
$\begin{equation}
h^*(s)=E(e^{-s(T+U)})=E(e^{-sT}e^{-sU})=E(e^{-sT})E(e^{-sU})=f^*(s)g^*(s)
\end{equation}$
\\
Asymptotic:
$\begin{equation}
    f^*(s) \sim \frac{A}{s}
\end{equation}$
As $s \downarrow 0$, as $t \rightarrow \infty$:
$\begin{equation}
    f(t) \longrightarrow A
\end{equation}$

\\
Multiple ($S_n$) dist:
$\begin{equation}
    f_n^*(s) = f^*(s)^n
\end{equation}$

# Metrics

$\begin{equation}
    Accurancy = \frac{TP+TN}{n}
\end{equation}$
e.g. how many of all predictions were right?

$\begin{equation}
    Precision = \frac{TP}{TP+FP}
\end{equation}$
e.g. how many of the predicted truths were right?

$\begin{equation}
    Recall = \frac{TP}{TP+FN}
\end{equation}$
e.g. how many of truths were predicted right?

$\begin{equation}
    F1 = 2\frac{Precision * Recall}{Precision + Recall}
    
\end{equation}$
e.g. mix of precision and recall
