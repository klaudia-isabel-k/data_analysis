```python
np.dot(sample_matrix_1, sample_matrix_2) == sample_matrix_1 @ sample_matrix_2
```

```python
dates = pd.date_range(datetime(2019,1,1), datetime(2023, 12, 31), freq='B')

df = pd.DataFrame({
    'date': dates,
    'return': [(np.random.normal(1.1, 0.01, 1) 
                if d.month in [6] 
                else 
                    (np.random.normal(1, 0.02, 1) if d.month in [9]
                        else np.random.normal(0.98, 0.01, 1))
               )[0] for d in dates]
}  
)
```

```python
df = df.set_index('date').resample('D').ffill()
df['value'] = np.cumprod(df['return']) * 10
df['year'] = df.index.year
df.index = df.index.map(lambda t: t.replace(year=2000, month=t.month, day=t.day))
df.reset_index(inplace=True)
```

```python
df_plot = df.pivot(index='date', columns='year', values='value')
```

```python
df_plot_scaled = df_plot / df_plot.iloc[0, ]
```

```python
df_plot_scaled.plot()
```

## Sin Wave

```python
t = np.linspace(0, 2 * np.pi, 100)
```

```python
y = np.sin(t)
```

```python
plt.plot(t, y)
plt.title('Sine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

## Normal Distribution Histogram

Objective: Plot the histogram of 100 random numbers sampled from a normal distribution. Label the x-axis as 'Value' and the y-axis as 'Frequency'. Give the plot a title.

```python
normal_sample = np.random.normal(0, 1, 100)
```

```python
plt.hist(normal_sample, bins=10)
plt.title('Normal Distribution Histogram (mean=0, std=1)')
plt.ylabel('Frequency')
plt.xlabel('Value')
plt.grid(True)
plt.show()
```
