```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0.05, 5, num=10)
y = np.log(x)
plt.plot(x, y, label='log')
plt.plot(x, [scipy.stats.norm.cdf(v, loc=0, scale=1) for v in y], label='CDR')
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = np.linspace(-2, 2, num=1000)
y = np.exp(x)
plt.plot(x, y, label='exp')
plt.grid(True)
plt.legend()
plt.show()
```
